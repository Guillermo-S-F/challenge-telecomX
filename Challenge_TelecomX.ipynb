```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json"
```

```python
df = pd.read_json(url)
```

```python
df.head()
```

## Expandir columna 'customer'


```python
customer_df = pd.json_normalize(df['customer'])
df = pd.concat([df, customer_df], axis=1)
df = df.drop('customer', axis=1)
print("Expanded 'customer' column and updated DataFrame.")
df.head()
```

## Expandir columna 'phone'



```python
phone_df = pd.json_normalize(df['phone'])
df = pd.concat([df, phone_df], axis=1)
df = df.drop('phone', axis=1)
print("Expanded 'phone' column and updated DataFrame.")
df.head()
```

## Expandir columna 'internet'



```python
internet_df = pd.json_normalize(df['internet'])
df = pd.concat([df, internet_df], axis=1)
df = df.drop('internet', axis=1)
print("Expanded 'internet' column and updated DataFrame.")
df.head()
```

## Expandir columna 'account'


```python
account_df = pd.json_normalize(df['account'])
df = pd.concat([df, account_df], axis=1)
df = df.drop('account', axis=1)
print("Expanded 'account' column and updated DataFrame.")
df.head()
```

## Mostrar DataFrame actualizado



```python
df.head()
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].replace({'Yes': 1, 'No': 0})

pd.set_option('display.max_columns', None)
print("Replaced 'Yes' with 1 and 'No' with 0 in appropriate columns.")
df.head()
```

```python
import numpy as np

df['Charges.Monthly'] = pd.to_numeric(df['Charges.Monthly'], errors='coerce').round(2)
df['Charges.Total'] = pd.to_numeric(df['Charges.Total'], errors='coerce').round(2)

print("Converted 'Charges.Monthly' and 'Charges.Total' columns to numeric with two decimal places.")
df.head()
```

```python
df.info()
```

## Anlisis Descriptivo de Datos


```python
print("### Estadsticas Descriptivas para Columnas Numricas:\n")
display(df.describe())

print("\n### Distribucin de Columnas Categricas:\n")
for col in df.select_dtypes(include='object').columns:
    print(f"\nColumna: {col}")
    display(df[col].value_counts())
```

## Preparar y Calcular Proporciones de Churn




```python
df['Churn'] = df['Churn'].replace('', np.nan)
df['Churn'] = pd.to_numeric(df['Churn'], errors='coerce')
df['Churn'] = df['Churn'].fillna(0).astype(int)

churn_proportions = df['Churn'].value_counts(normalize=True)

print("Converted 'Churn' column to integer type and handled problematic values.")
print("### Proporcin de clientes que permanecieron (0) y se dieron de baja (1):")
display(churn_proportions)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Bar Chart for Churn Proportions
sns.barplot(x=churn_proportions.index, y=churn_proportions.values, ax=axes[0], palette='viridis')
axes[0].set_title('Proportion of Customer Churn', fontsize=14)
axes[0].set_xlabel('Churn (0 = No, 1 = Yes)', fontsize=12)
axes[0].set_ylabel('Proportion', fontsize=12)
axes[0].set_xticks([0, 1])
axes[0].set_xticklabels(['No Churn', 'Churn'])
for index, value in enumerate(churn_proportions.values):
    axes[0].text(index, value + 0.01, f'{value:.2%}', ha='center', va='bottom', fontsize=10)

# Pie Chart for Churn Proportions
axes[1].pie(churn_proportions.values, labels=['No Churn', 'Churn'], autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'salmon'], wedgeprops={'edgecolor': 'black'})
axes[1].set_title('Distribution of Customer Churn', fontsize=14)
axes[1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.tight_layout()
plt.show()
```

```python
object_cols = df.select_dtypes(include='object').columns
categorical_cols = [col for col in object_cols if col not in ['customerID', 'Churn']]

print("Identified categorical columns (excluding 'customerID' and 'Churn'):")
print(categorical_cols)
```

## Visualizar Proporciones de Churn por Columna Categrica


```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    churn_by_category = df.groupby(col)['Churn'].value_counts(normalize=True).unstack().fillna(0)

    # Ensure the '1' column (churned) exists for plotting
    if 1 in churn_by_category.columns:
        sns.barplot(x=churn_by_category.index, y=churn_by_category[1], palette='viridis')
        plt.title(f'Churn Proportion by {col}', fontsize=14)
        plt.xlabel(col, fontsize=12)
        plt.ylabel('Proportion of Churn (1)', fontsize=12)
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
    else:
        print(f"No churn (value 1) observed in column '{col}'. Skipping plot.")

print("Visualized churn proportions for all identified categorical columns.")
```

## Resumir Patrones de Churn por Categora



## Resumen de Patrones de Churn por Categora

### Observaciones Clave:

*   **gender (gnero)**: Las tasas de churn entre hombres y mujeres son muy similares, lo que sugiere que el gnero por s solo no es un factor significativo en la decisin de un cliente de darse de baja.

*   **MultipleLines (mltiples lneas)**: Los clientes con servicio de mltiples lneas (`Yes`) muestran una tasa de churn ligeramente ms baja que aquellos sin mltiples lneas (`No`), y aquellos sin servicio telefnico (`No phone service`) tienen la menor tasa de churn, lo cual es de esperar ya que no tienen un servicio principal que cancelar.

*   **InternetService (servicio de Internet)**:
    *   **Fiber optic (fibra ptica)**: Esta categora presenta la tasa de churn ms alta, lo que podra indicar problemas de calidad de servicio, precio o insatisfaccin con el plan de fibra ptica.
    *   **DSL**: Los clientes con DSL tienen una tasa de churn moderada.
    *   **No**: Los clientes sin servicio de internet tienen la tasa de churn ms baja, lo cual es lgico ya que no hay un servicio de internet que puedan cancelar.

*   **OnlineSecurity (seguridad en lnea)**:
    *   Los clientes que **no** tienen seguridad en lnea (`No`) tienen una tasa de churn significativamente ms alta en comparacin con aquellos que s la tienen (`Yes`) o que no tienen servicio de internet. Esto sugiere que la seguridad en lnea puede ser un factor de retencin importante.

*   **OnlineBackup (copia de seguridad en lnea)**:
    *   Similar a la seguridad en lnea, los clientes sin copia de seguridad en lnea (`No`) muestran una tasa de churn ms alta, indicando que este servicio adicional tambin contribuye a la retencin.

*   **DeviceProtection (proteccin de dispositivos)**:
    *   Los clientes que **no** tienen proteccin de dispositivos (`No`) tienen una tasa de churn ms alta que aquellos que s la tienen, reforzando la idea de que los servicios adicionales son importantes para la retencin.

*   **TechSupport (soporte tcnico)**:
    *   Los clientes que **no** tienen soporte tcnico (`No`) presentan una tasa de churn considerablemente ms alta. Esto subraya la importancia de un buen soporte tcnico para la satisfaccin del cliente y la retencin.

*   **StreamingTV (TV en streaming) y StreamingMovies (pelculas en streaming)**:
    *   Los clientes que **no** utilizan estos servicios de streaming (`No`) tienen tasas de churn ms altas que aquellos que s los utilizan (`Yes`). Esto sugiere que la oferta de entretenimiento puede ser un factor de retencin.

*   **Contract (tipo de contrato)**:
    *   **Month-to-month (mes a mes)**: Los clientes con contratos mes a mes tienen la tasa de churn ms alta con diferencia. Esto es esperado, ya que estos contratos ofrecen mayor flexibilidad para que los clientes se vayan.
    *   **One year (un ao)** y **Two year (dos aos)**: Estos contratos tienen tasas de churn mucho ms bajas, siendo los contratos de dos aos los que presentan la menor tasa de churn, lo que indica que los compromisos a largo plazo son un fuerte factor de retencin.

*   **PaymentMethod (mtodo de pago)**:
    *   **Electronic check (cheque electrnico)**: Los clientes que utilizan el cheque electrnico tienen la tasa de churn ms alta.
    *   **Bank transfer (automatic)** y **Credit card (automatic)**: Estos mtodos de pago automticos muestran tasas de churn considerablemente ms bajas, lo que podra indicar una mayor comodidad o lealtad del cliente.
    *   **Mailed check (cheque por correo)**: Presenta una tasa de churn intermedia.

### Hiptesis y Prximos Pasos:

*   **Servicios de Valor Aadido**: La ausencia de servicios como seguridad en lnea, copia de seguridad, proteccin de dispositivos y soporte tcnico est fuertemente correlacionada con mayores tasas de churn. Ofrecer y promover estos servicios podra mejorar la retencin.
*   **Tipo de Contrato**: Los contratos a largo plazo (uno y dos aos) son un fuerte indicador de menor churn. Podra ser beneficioso incentivar a los clientes a optar por estos contratos.
*   **Mtodo de Pago**: El cheque electrnico como mtodo de pago est asociado con un churn elevado. Se podra investigar por qu los clientes que usan este mtodo estn ms propensos a irse.
*   **Servicio de Fibra ptica**: La alta tasa de churn en clientes con fibra ptica es una seal de alerta que requiere una investigacin ms profunda sobre posibles problemas de rendimiento, precios o expectativas no cumplidas.

## Calcular Nmero de Servicios Contratados



```python
service_cols_to_convert = ['MultipleLines', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

print("Defined service columns for conversion:")
print(service_cols_to_convert)
```

```python
for col in service_cols_to_convert:
    df[col] = df[col].replace({'No phone service': 0, 'No internet service': 0, 'Yes': 1, 'No': 0})
    df[col] = df[col].astype(int)

print("Converted service-related columns to numerical (0/1) and cast to integer type.")
df.head()
```

```python
all_service_cols = service_cols_to_convert + ['PhoneService']
df['Nmero de Servicios Contratados'] = df[all_service_cols].sum(axis=1)

print("Created 'Nmero de Servicios Contratados' column.")
df.head()
```
